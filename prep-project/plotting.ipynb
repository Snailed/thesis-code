# Misc plotting from mcmc_stats

```python
txt = ""
with open("results/2025-01-15 14:00:07.656845/mcmc_stats/Exp1WBNN_4_deep_8_wide", "r") as f:
    txt = f.read()
Array = np.array
d = eval(txt)
dicts = []
for chain in range(8):
    for i in range(len(d["i"][chain])):
        v = {
            "chain": chain
        }
        for key in d.keys():
            v[key] = d[key][chain][i]
        dicts.append(v)
df = pd.DataFrame(dicts)
```

```python
sns.lineplot(data=df, x="i", y="mean_accept_prob", hue="chain")
```
